# Notebook 08 – Role Specific Views

## Why Role Specific Views?

Different healthcare users require different levels of access to patient information.

For example:

- Clinicians require complete patient records for treatment.
- Data Analysts require anonymized data for reporting and analytics.
- Auditors require access for compliance and governance purposes.

This notebook creates role-specific SQL views that provide the appropriate level of data access while protecting sensitive healthcare information.

### Objective

The objective of this notebook is to create secure role-specific views for different healthcare users by exposing only the information required for their responsibilities.

### Input Tables

This notebook reads the following governed tables:

- gold.patient_masked
- gold.patient_cohort_masked
- gold.readmission_predictions_masked

### Output

This notebook creates role-specific SQL views including:

- clinician_patient_view
- analyst_patient_view
- auditor_patient_view
- clinician_prediction_view
- analyst_prediction_view
- auditor_prediction_view


### 1. Create Clinician Views

Clinicians require complete patient information to support diagnosis and treatment.

These views expose patient-level healthcare data while maintaining governance controls defined in the SQL Governance layer.

```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.clinician_view AS

SELECT
    *
FROM db_healthcare_kl.gold.readmission_predictions;
```

### 2. Create Analyst Views

Data analysts require healthcare data for reporting and business intelligence.

Personally Identifiable Information (PII) is masked while retaining clinical and operational information required for analytics.

```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.analyst_view AS

SELECT

    SHA2(person_id,256) AS person_id,

    total_visits,
    max_duration_days,
    avg_duration_days,
    age,
    gender_concept_id,
    diagnosis_count,
    medication_count,
    previous_admissions,
    visit_type,
    Cohort_name,
    readmission_30day_flag,
    prediction

FROM db_healthcare_kl.gold.readmission_predictions;
```

### 3. Create Auditor Views

Auditors require secure access to healthcare records for compliance verification and governance reviews.

The auditor views provide controlled access while maintaining data integrity and traceability.

```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.auditor_view AS

SELECT

    processed_timestamp,
    prediction,
    readmission_30day_flag,
    Cohort_name

FROM db_healthcare_kl.gold.readmission_predictions;
```
