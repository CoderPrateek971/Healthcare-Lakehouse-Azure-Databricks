```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.clinician_view AS

SELECT
    *
FROM db_healthcare_kl.gold.readmission_predictions;
```

```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.analyst_view AS

SELECT

    SHA2(person_id,256) AS person_id,

    total_visits,
    max_duration_days,
    avg_duration_days,
    age,
    gender_concept_id,
    diagnosis_count,
    medication_count,
    previous_admissions,
    visit_type,
    Cohort_name,
    readmission_30day_flag,
    prediction

FROM db_healthcare_kl.gold.readmission_predictions;
```

```python
%sql

CREATE OR REPLACE VIEW db_healthcare_kl.gold.auditor_view AS

SELECT

    processed_timestamp,
    prediction,
    readmission_30day_flag,
    Cohort_name

FROM db_healthcare_kl.gold.readmission_predictions;
```
